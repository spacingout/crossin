```python
# coding=utf-8
import pandas as pd
from pandas import Series
from matplotlib import pyplot as plt
```

```python
# 1.读取 CSV 格式文件“task_2_lianjia_data.csv”，将数据命名为 lianjia_data，并选取前部数据查看

f = open("task_2_lianjia_data.csv")
lianjia_data = pd.read_csv(f)
lianjia_data.head()
```

```python
# 2.利用 groupby 函数，以【面积大小(m²)】字段进行 groupby，统计各个面积下的【房源数量】情况
gb_count = [('房源数量','count')]
df2 = lianjia_data.groupby('面积大小（㎡）')['简介（方式·小区名 户型 朝向）'].agg(gb_count)
```

```python
# 3.将第(2)问所统计数据，作折线图和条形图，并得出相关结论(阅读参考文章 5、6)，提示:先提取第(2)问 所得数据表的 index 索引，饼转化成 list;再提取【房源数量】的值转成 list(阅读参考文章 7、8);
mianji = df2.index.values
print(mianji)
shuliang = df2['房源数量'].values.tolist()
print(shuliang)


# 绘制图表
plt.subplot(2,1,1)
plt.plot(mianji,shuliang)
plt.subplot(2,1,2)
plt.bar(mianji,shuliang)

# 结论：整体来看，上海市的待租户型以小户型为主，特别集中于30-50平米左右。
```

```python
# 4. 利用 groupby 函数，以【区】字段进行 groupby，统计各个上海各个“区”所属的【房源数量】情况;
gb_count = [('房源数量','count')]
df_qu = lianjia_data.groupby('区')['简介（方式·小区名 户型 朝向）'].agg(gb_count)
df_qu
```

```python

# 5. 将第(4)问所统计数据，作条形图，并得出相关结论;
qu = df_qu.index.values
print(qu)
shuliang2 = df_qu['房源数量'].values.tolist()
print(shuliang2)

plt.bar(qu,shuliang2)

# 嘉定、奉贤、青浦房源较少，徐汇、浦东、静安区房源较多
```

```python
# 6.利用 groupby 函数，以【楼层类型】字段进行 groupby，统计各个楼层所属的【房源数量】情况;
gb_count = [('房源数量','count')]
df_floor = lianjia_data.groupby('楼层类型')['简介（方式·小区名 户型 朝向）'].agg(gb_count)

floor = df_floor.index.values
shuliang3 = df_floor['房源数量'].values.tolist()

fig = plt.figure()
plt.pie(shuliang3,labels=floor,autopct='%1.2f%%') #画饼图（数据，数据对应的标签，百分数保留两位小数点）
plt.title("Pie chart")
plt.show()

# 结论：待租房源中高楼层占比最多，达到了近一半，仍有少量房源为地下室
```
